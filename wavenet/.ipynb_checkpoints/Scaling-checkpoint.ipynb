```python
"""PreProcess Data"""
# 1. Mu encode the data
mu_encoder = torchaudio.transforms.MuLawEncoding(quantization_channels=256)
mu_decoder = torchaudio.transforms.MuLawDecoding(quantization_channels=256)
```

```python
sample = torch.Tensor(sound_pickle[0])
sample_mu_encoded = mu_encoder(sample)
```

```python
plt.plot(sample_mu_encoded.numpy()[:, 0])
plt.show()
```

```python
plt.plot(sample[:, 0])
plt.show()
```

```python
rescaled = sample / 100
plt.plot(rescaled[:, 0])
plt.show()
```

```python
rescaled_encoded = torch.Tensor(rescaled)
rescaled_encoded = mu_encoder(rescaled_encoded)
plt.plot(rescaled_encoded.numpy()[:,0])
plt.show()
```

```python
decoded = mu_decoder(rescaled_encoded) * 100
torch.sum(sample - decoded)
```

```python
s = sample.numpy()[:, 0]
d = decoded.numpy()[:, 0]
plt.plot(d)
plt.show()
plt.plot(s)
plt.show()
```
